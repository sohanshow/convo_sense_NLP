```python
import pickle
```

```python
bar = 100000
```

```python
# wordlist file path
wordlist_path= "/content/drive/MyDrive/Colab Notebooks/Upwork-notebooks/Detect-OOV/google_freq_dump.json"
```

```python
import json

with open(wordlist_path) as json_file:
    word_freq_allwords = json.load(json_file)
```

```python
unique = []
common = []


for x in word_freq_allwords.keys():
  if word_freq_allwords[x] > bar:
    common.append(x)
  else:
    unique.append(x)
```

```python
f = open('/content/drive/MyDrive/Colab Notebooks/Upwork-notebooks/Detect-OOV/unique_bar100000.txt', 'w')
temp =';;'.join(unique)

f.write(temp)
f.close()
```
